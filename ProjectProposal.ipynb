Project Title – Traffic Accidents in Great Britain

Project Plan – I will use data that contains information on traffic accidents in Great Britain. The data were collected from 2005-2014.  It contains accidents from every crash that was reported to the police on public roads.  The data come from a form that is filled out by the police for each accident and sent to the Department of Transportation.  The goal for the project is to explore all of the variables and find the ones that are the most relevant for predicting whether or not there were any casualties.  I will also compare several different methods of classification.

Model Evaluation – I will use logistic regression, k nearest neighbors, decision trees, naive bayes, and support vector machine.  I will use ROC/AUC to evaluate the model's performance. 

Data Sources – I found the data at data.gov.uk under road safety data from the Department of Transportation.  There are over 40 variables that are grouped into three different datasets: accident circumstances, vehicle, and casualty.  The accident circumstances dataset includes 35 variables.  Some of them are weather conditions, light conditions, number of vehicles, pedestrian crossing, date, time, and road surface conditions.  The vehicle dataset has over 20 variables including age of vehicle, vehicle type, vehicle maneuver, and skidding and overturning.  The casualty dataset includes variables from the previous two datasets along with age of casualty, sex of casualty, car passenger, bus passenger, and severity.  The three data sets are linked by the variable accident index. The vehicles and casualties datasets are also linked by vehicle reference.
